## 1. How many customers did we gain and lose this period?  

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import sum

df = spark.read.table("subscription_pipeline.gold.fact_subscription_revenue")

window_12 = Window.orderBy("revenue_date").rowsBetween(-11, 0)

rolling = df.withColumn(
"rolling_12m_revenue",
sum("revenue_gbp").over(window_12)
)

rolling.display()
```
