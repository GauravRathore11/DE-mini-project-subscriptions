```python
from pyspark.sql.functions import col, current_timestamp

df = spark.read.table("subscription_pipeline.silver.fx_rates")

dim_fx = df.select(
col("currency_code"),
col("fx_rate_to_gbp"),
col("effective_date")
).dropDuplicates(["currency_code", "effective_date"]) \
.withColumn("dw_created_at", current_timestamp())

dim_fx.write \
.mode("overwrite") \
.format("delta") \
.saveAsTable("subscription_pipeline.gold.dim_fx_rate")

print("dim_fx_rate created")
```
