```python
from pyspark.sql.functions import *

employees = spark.read.table("subscription_pipeline.silver.employees")

dim_employee = employees.select(
    "employee_id",
    "employee_name",
    "role",
    "region",
    "is_active_flag"
).dropDuplicates()

dim_employee.write.mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("subscription_pipeline.gold.dim_employee")
```
