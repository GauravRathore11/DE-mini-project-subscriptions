```python
from pyspark.sql.functions import col, current_timestamp

df = spark.read.table("subscription_pipeline.silver.employees")

dim_employee = df.select(
col("employee_id"),
col("employee_name"),
col("role"),
col("region"),
col("hire_date"),
col("is_active_flag").cast("boolean").alias("is_active")
).dropDuplicates(["employee_id"]) \
.withColumn("dw_created_at", current_timestamp())

dim_employee.write \
.mode("overwrite") \
.format("delta") \
.saveAsTable("subscription_pipeline.gold.dim_employee")

print("dim_employee created")
```
