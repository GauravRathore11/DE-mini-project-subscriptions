```python
from pyspark.sql.functions import col, current_timestamp

df = spark.read.table("subscription_pipeline.silver.countries")

dim_country = df.select(
col("country_code"),
col("country_name"),
col("currency_code")
).dropDuplicates(["country_code"]) \
.withColumn("dw_created_at", current_timestamp())

dim_country.write \
.mode("overwrite") \
.format("delta") \
.saveAsTable("subscription_pipeline.gold.dim_country")

print("dim_country created")
```
