```python
from pyspark.sql.functions import *

countries = spark.read.table("subscription_pipeline.silver.countries")

dim_country = countries.select(
    "country_code",
    "country_name",
    "currency_code"
).dropDuplicates()

dim_country.write.mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("subscription_pipeline.gold.dim_country")
```
