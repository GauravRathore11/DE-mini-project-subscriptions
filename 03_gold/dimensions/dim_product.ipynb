```python
from pyspark.sql.functions import col, current_timestamp

df = spark.read.table("subscription_pipeline.silver.products")

dim_product = df.select(
col("product_id"),
col("product_name"),
col("plan_name"),
col("billing_cycle"),
col("list_price"),
col("is_active").cast("boolean"),
col("created_date")
).dropDuplicates(["product_id"]) \
.withColumn("dw_created_at", current_timestamp())

dim_product.write \
.mode("overwrite") \
.format("delta") \
.saveAsTable("subscription_pipeline.gold.dim_product")

print("dim_product created")
```
