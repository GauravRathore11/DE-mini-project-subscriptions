```python
from pyspark.sql.functions import *

products = spark.read.table("subscription_pipeline.silver.products")

dim_product = products.select(
    "product_id",
    "product_name",
    "plan_name",
    "billing_cycle",
    "list_price",
    "is_active"
).dropDuplicates()

dim_product.write.mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("subscription_pipeline.gold.dim_product")
```
