```python
from pyspark.sql.functions import *

customers = spark.read.table("subscription_pipeline.silver.customers")

dim_customer = customers.select(
    "customer_id",
    "customer_name",
    "country",
    "industry_type",
    "account_created_date",
    "is_active"
).dropDuplicates()

# Optional: add customer_type
dim_customer = dim_customer.withColumn(
    "customer_type",
    when(col("is_active") == 1, "active").otherwise("inactive")
)

dim_customer.write.mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("subscription_pipeline.gold.dim_customer")
```
