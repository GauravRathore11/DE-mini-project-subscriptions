```python
from pyspark.sql.functions import col, current_timestamp

df = spark.read.table("subscription_pipeline.silver.customers")

dim_customer = df.select(
col("customer_id"),
col("customer_name"),
col("country"),
col("industry_type"),
col("account_created_date"),
col("is_active").cast("boolean")
).dropDuplicates(["customer_id"]) \
.withColumn("dw_created_at", current_timestamp())

dim_customer.write \
.mode("overwrite") \
.format("delta") \
.saveAsTable("subscription_pipeline.gold.dim_customer")

print("dim_customer created")
```
